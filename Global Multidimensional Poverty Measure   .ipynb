```python
import pandas as pd
df=pd.read_csv("MPI_data.csv")
display(df.head())
```

```python
df.drop(columns='Series Code',inplace=True)
```

```python
df.rename(columns={'Series Name':'Indicators','2014 [YR2014]':'2014','2015 [YR2015]':'2015','2016 [YR2016]':'2016','2017 [YR2017]':'2017','2018 [YR2018]':'2018','2019 [YR2019]':'2019','2020 [YR2020]':'2020','2021 [YR2021]':'2021','2022 [YR2022]':'2022','2023 [YR2023]':'2023'},inplace=True)
```

```python
df['Indicators']=df['Indicators'].str.replace('Adjusted net enrollment rate, primary (% of primary school age children)','enrollment of children in primary education')
df['Indicators']=df['Indicators'].str.replace('Children out of school (% of primary school age)','children out of school')
df['Indicators']=df['Indicators'].str.replace('Current education expenditure, total (% of total expenditure in public institutions','current education expenditure')
df['Indicators']=df['Indicators'].str.replace("Educational attainment, at least Bachelor's or equivalent, population 25+, total (%) (cumulative)","Educational attainment(25+)")
df['Indicators']=df['Indicators'].str.replace('Government expenditure on education, total (% of GDP)','government expenditure on education')
df['Indicators']=df['Indicators'].str.replace('Crop production index (2014-2016 = 100)','crop production index')
df['Indicators']=df['Indicators'].str.replace('Access to electricity (% of population)','access to electricity')
df['Indicators']=df['Indicators'].str.replace('Population living in slums (% of urban population)','slum_population')
df['Indicators']=df['Indicators'].str.replace('Agricultural land (% of land area)','agricultural land')
df['Indicators']=df['Indicators'].str.replace('Food production index (2014-2016 = 100)','food production index')
df['Indicators']=df['Indicators'].str.replace('Literacy rate, youth total (% of people ages 15-24)','literacy rate')
df['Indicators']=df['Indicators'].str.replace('Rural population growth (annual %)','rural population growth')
df['Indicators']=df['Indicators'].str.replace('Urban population growth (annual %)','urban population growth')
df['Indicators']=df['Indicators'].str.replace('Water productivity, total (constant 2015 US$ GDP per cubic meter of total freshwater withdrawal)','water productivity')
df['Indicators']=df['Indicators'].str.replace('Learning poverty: Share of Children at the End-of-Primary age below minimum reading proficiency adjusted by Out-of-School Children (%)','learning poverty')
```

```python
import numpy as np
df.replace('..',np.nan,inplace=True)
```

```python
display(df.head(5))
```

```python
x=pd.DataFrame({'Country Name': df['Country Name'],'Country Code':df['Country Code'],'Indicators':df['Indicators']})
display(x)
```

```python
y=df.drop(columns=['Country Name','Country Code','Indicators'])
y = y.apply(pd.to_numeric, errors='coerce')
display(y)
y['2020'].info
```

```python
def fillna_with_row_mean(row):
    row_mean = row.mean()
    return row.fillna(row_mean)
y = y.apply(lambda row: fillna_with_row_mean(row), axis=1)
display(y)
```

```python
data=pd.DataFrame({'Country Name': x['Country Name'],'Country Code':x['Country Code'],'Indicators':x['Indicators'],'2014':y['2014'],'2015':y['2015'],'2016':y['2016'],'2017':y['2017'],'2018':y['2018'],'2019':y['2019'],'2020':y['2020'],'2021':y['2021'],'2022':y['2022'],'2023':y['2023']})
display(data)
```
